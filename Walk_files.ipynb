```python
import os
import re


sep = os.sep # equals / or \ respectively on UNIX and Windows

def get_dict(path):
    for root, dirs, files in os.walk(path):#
        foldername = root.split(sep)[-1] #isolate last part of path
        directories = {}

        for file in files:
            if(re.match(".*coSMIC_all_P[0-9][0-9]_C[0-9]\\.wmv", root+file)):
                attributes = {}
                attributes["path"] = root + file
                attributes["valid"] = True

                directories[file] = attributes

        for d in dirs:
            contents = {}
            directories[d] = get_dict(root+d)
            
        return directories
```

```python
# Convert to dataframe such that each toplevel entry (1,2,3,4....) yields one line
# Row headers will be e.g. "video.1.valid"
# Missing values will be None
d = direc

pairs = [key for key in d.keys()]

un_id = []
def acc_rec(node, prefix):
    for key, value in node.items():
        if(isinstance(value, dict)):
            new =[]
            new.extend(prefix)
            new.append(key)
            acc_rec(value, new)
        else:
            new = []
            new.extend(prefix)
            new.append(key)
            un_id.append(new)
            
def accumulate_names(node):
    for value in node.values():
        acc_rec(value, [])
        
accumulate_names(direc)
un_id = [list(x) for x in set(tuple(i) for i in un_id)]
un_id.sort()
```

```python
import numpy as np
import pydot    

class Dotstring:
    cid = 0
    @staticmethod
    def get_id():
        Dotstring.cid += 1
        return str(Dotstring.cid)

    def visit1(node, parent_key, parent_id, omit_leaves = True):
        """ Assembles a dotstring given a nested dictionary.
            Nodes are regarded to have different values also if they share the same values.
            In other words multiple linkages will be avoided.
        """
        dotstring = ''
        if(not parent_key == None):
            for k,v in node.items():
                cid = Dotstring.get_id()
                dotstring +=  cid + ' [label ="' + str(k) +'"];\n'#Add current node (parent has been added)
                dotstring +=  parent_id + ' -- ' + cid + "\n" #link parent id to current id
                if(isinstance(v, dict)):
                    dotstring += Dotstring.visit1(v, k, cid)
                
                if(omit_leaves):
                    continue
                    
                if(isinstance(v, str) or isinstance(v, int) or isinstance(v, float)):
                    leave_id = Dotstring.get_id()
                    dotstring +=  leave_id + ' [label ="' + str(v) +'"];\n'#Add current node (parent has been added)
                    dotstring +=  cid + ' -- ' + leave_id + "\n" #link parent id to current id

        else:
            assert parent_key == None
            cid = Dotstring.get_id()
            dotstring +=  cid + ' [label ="' + list(node.keys())[0] +'"];\n'#Add current node (parent has been added)
            for k,v in node.items():
                dotstring += Dotstring.visit1(v,"Dinner",cid)
        return dotstring


    def visit(node, parent_key):
        """ Assembles a dotstring given a nested dictionary. Multiple linkages are possible if a value appears more then once"""

        dotstring = ''
        for k,v in node.items():
            if(not parent_key == None):#If there is a parent draw connections from parent to child
                dotstring +=   parent_key +" -- " + k + ";\n"
            if(not isinstance(v, dict)):#If node is a leave draw connections to leaves
                if(isinstance(v, str) or isinstance(v, int) or isinstance(v, float)):# if its a scalar
                    dotstring += k +" -- " + str(v) + ";\n"
                elif(isinstance(v, list)):
                    dotstring += k +" -- " + str(v[0:3]) + ";\n"
            if(isinstance(v, dict)):#Continue with subtree
                dotstring += Dotstring.visit(v, k)
        return dotstring

    def assemble_dotstring(dictionary, mode = "by_id"):
        if(mode == "by_name"):
            dotstring = "graph {"
            dotstring += Dotstring.visit(dictionary, None)
            dotstring += "}"
        if(mode == "by_id"):
            dotstring = "strict graph {"
            string = Dotstring.visit1(dictionary, None, 0)
            dotstring += string
            dotstring += "}"

        return dotstring
```
